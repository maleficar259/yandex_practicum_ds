### –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ä–∞—Å—á–µ—Ç–∞ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ–±–æ–≥–∞—â–µ–Ω–∏—è

–ù–∞ –¥–∞–Ω–Ω–æ–º —ç—Ç–∞–ø–µ –Ω–∞–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å, –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –ª–∏ —Ä–∞—Å—á–∏—Ç–∞–Ω–∞ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –æ–±–æ–≥–∞—â–µ–Ω–∏—è –Ω–∞ –ø—Ä–∏–º–µ—Ä–µ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏ –∏ –ø–µ—Ä–≤–æ–≥–æ —ç—Ç–∞–ø–∞ –æ–±—Ä–∞–±–æ—Ç–∫–∏ —Ä—É–¥—ã (—Ñ–ª–æ—Ç–∞—Ü–∏–∏), –∞ —Ç–∞–∫–∂–µ –æ—Ü–µ–Ω–∏—Ç—å MAE –∏–º–µ—é—â–∏—Ö—Å—è –∏ —Ä–∞—Å—á–µ—Ç–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π.

```python
# —Å–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é, —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞—é—â—É—é —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –æ–±–æ–≥–∞—â–µ–Ω–∏—è

def recovery_calc(row):
    try:
        recovery = ((row['rougher.output.concentrate_au'] * (row['rougher.input.feed_au'] - row['rougher.output.tail_au'])) \
                / (row['rougher.input.feed_au'] * (row['rougher.output.concentrate_au'] - row['rougher.output.tail_au']))) * 100
        return recovery
    except:
        return None
```

```python
data_train['rougher.output.recovery_calc'] = data_train.apply(recovery_calc, axis=1)
data_train[data_train['rougher.output.recovery_calc'].isna()]
```

```python
# –Ω–∞–π–¥–µ–º MAE –º–µ–∂–¥—É —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –∏ –∏–º–µ—é—â–∏–º—Å—è –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ –≤ —Ç–∞–±–ª–∏—Ü–µ
mae = mean_absolute_error(data_train['rougher.output.recovery'], data_train['rougher.output.recovery_calc'])
mae
```

–¢–∞–∫–∂–µ —Å—Ç–æ–∏—Ç –æ—Ç–º–µ—Ç–∏—Ç—å, —á—Ç–æ –µ—Å—Ç—å —Ç–∞–∫–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ –∏–º–µ—é—Ç –æ—á–µ–Ω—å —Å–∏–ª—å–Ω—É—é –ø–æ–ø–∞—Ä–Ω—É—é –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é –º–µ–∂–¥—É —Å–æ–±–æ–π (–∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ü–∏—Ä—Å–æ–Ω–∞ >0.9). –î–ª—è –∫–∞–∂–¥–æ–π –≥—Ä—É–ø–ø—ã —Å–∏–ª—å–Ω–æ —Å–∫–æ—Ä—Ä–µ–ª–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —Å—Ç–æ–∏—Ç –æ—Å—Ç–∞–≤–∏—Ç—å —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω, –∞ –æ—Å—Ç–∞–ª—å–Ω—ã–µ –∏—Å–∫–ª—é—á–∏—Ç—å.

–î–ª—è –Ω–∞—á–∞–ª–∞ –æ–±—Ä–∞–±–æ—Ç–∞–µ–º –≤—ã–±—Ä–æ—Å—ã. –£–¥–∞–ª–∏–º –∏–∑ –∫–∞–∂–¥–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞ –¥–∞–Ω–Ω—ã–µ, –∫–æ—Ç–æ—Ä—ã–µ –±–æ–ª–µ–µ —á–µ–º –Ω–∞ 3 —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã—Ö –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –±–æ–ª—å—à–µ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è –æ—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–µ.

```python
# —Å–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –æ—Ç—Å–µ—á–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö —Å–≤–µ—Ä—Ö +- 3 —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã—Ö –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è
def std3_cut(data):
    data_columns = list(data.columns)
    
    for i in data_columns:
        mean = data[i].mean()
        std = data[i].std()
        data[i] = data[i][(data[i] > (mean - std * 3)) & (data[i] < (mean + std * 3))]
    return data
```

```python
data_cut = std3_cut(data_cut)
data_train_cut = std3_cut(data_train_cut)
data_test = std3_cut(data_test)
```

```python
# –ø–æ–≤—Ç–æ—Ä–Ω–æ –ø—Ä–æ–≤–µ–¥–µ–º –ø—Ä–æ—Ü–µ–¥—É—Ä—É –∏–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏–∏, —á—Ç–æ–±—ã –∑–∞–ø–æ–ª–Ω–∏—Ç—å –ø–æ—è–≤–∏–≤—à–∏–µ—Å—è –ø–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è –≤—ã–±—Ä–æ—Å–æ–≤ –ø—Ä–æ–ø—É—Å–∫–∏
data_cut.interpolate(inplace=True)
data_train_cut.interpolate(inplace=True)
data_test.interpolate(inplace=True)
```

```python
# —Å–æ–∑–¥–∞–µ–º –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É
data_corr = data_cut.corr()
# –∑–∞–ø–æ–ª–Ω—è–µ–º –Ω—É–ª—è–º–∏ –≤—Å—ë, –Ω–∞—á–∏–Ω–∞—è —Å –¥–∏–∞–≥–æ–Ω–∞–ª–∏ –∏ –≤—ã—à–µ 
data_corr[:] = np.tril(data_corr.values, k=-1)
# –≤—ã–±–∏—Ä–∞–µ–º —Å—Ç—Ä–æ–∫–∏ —Ç–æ–ª—å–∫–æ —Å–æ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –ü–∏—Ä—Å–æ–Ω–∞ –±–æ–ª—å—à–µ –º–æ–¥—É–ª—è 0.9 –∏ != 1
data_corr = data_corr[((data_corr > 0.9) | (data_corr < -0.9)) & (data_corr < 1)]

data_corr.dropna(how='all', inplace=True)
data_corr.dropna(axis='columns', how='all', inplace=True)

data_corr
```

## –ú–æ–¥–µ–ª—å

–ù–∞ –¥–∞–Ω–Ω–æ–º —ç—Ç–∞–ø–µ –ø—Ä–æ–µ–∫—Ç–∞ –Ω–∞–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ:
- –ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –∏—Ç–æ–≥–æ–≤–æ–π sMAPE.
- –û–±—É—á–∏—Ç—å —Ä–∞–∑–Ω—ã–µ –º–æ–¥–µ–ª–∏ –∏ –æ—Ü–µ–Ω–∏—Ç–µ –∏—Ö –∫–∞—á–µ—Å—Ç–≤–æ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–µ–π. 
- –í—ã–±—Ä–∞—Ç—å –ª—É—á—à—É—é –º–æ–¥–µ–ª—å –∏ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –µ—ë –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ. 
- –û–ø–∏—Å–∞—Ç—å –≤—ã–≤–æ–¥—ã. 

–ù–∞—á–Ω—ë–º –ø–æ –ø–æ—Ä—è–¥–∫—É.

```python
# —Å–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ sMAPE
def smape_count(target, predictions):
    numerator = abs(target - predictions)
    denominator = ((abs(target) + abs(predictions)) / 2)
    smape = (numerator / denominator).mean() * 100
    return smape
```

```python
# —Ç–µ–ø–µ—Ä—å —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —Ä–∞—Å—Å—á–µ—Ç–∞ –∏—Ç–æ–≥–æ–≤–æ–≥–æ smape
def final_smape_count(target_1, predictions_1, target_2, predictions_2):
    smape_1 = smape_count(target_1, predictions_1)
    smape_2 = smape_count(target_2, predictions_2)
    final_smape = 0.25 * smape_1 + 0.75 * smape_2
    return final_smape
```
