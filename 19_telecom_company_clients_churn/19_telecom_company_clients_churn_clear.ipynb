```python
# –ø—Ä–æ–≤–µ–¥—ë–º Ordinal encoding –¥–ª—è –º–æ–¥–µ–ª–∏ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞

features_train_ord_encoded = features_train.copy()
features_test_ord_encoded = features_test.copy()

object_features_list = features_train.select_dtypes(include='object').columns.to_list()
print(object_features_list)
```

```python
# –¥–ª—è –º–æ–¥–µ–ª–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ - OHE –∏ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ

features_train_ohe_scaled = features_train.copy()
features_test_ohe_scaled = features_test.copy()
```

```python
col_transformer_ohe_scaler = make_column_transformer(
    (
        OneHotEncoder(drop='first', handle_unknown='ignore'),
        object_features_list
    ),
    (
        StandardScaler(),
        num_features_list
    ),
    remainder='passthrough',
    verbose_feature_names_out=False
)
```

    - –ü–æ–¥–∫–ª—é—á–µ–Ω –ª–∏ —É –∫–ª–∏–µ–Ω—Ç–∞ –∏–Ω—Ç–µ—Ä–Ω–µ—Ç.
    - –ü–æ–¥–∫–ª—é—á–µ–Ω –ª–∏ —É –∫–ª–∏–µ–Ω—Ç–∞ —Ç–µ–ª–µ—Ñ–æ–Ω.
    - –ü–æ–ª—å–∑—É–µ—Ç—Å—è –ª–∏ –∫–ª–∏–µ–Ω—Ç –ª—é–±—ã–º–∏ –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–º–∏ —É—Å–ª—É–≥–∞–º–∏.
- –ò—Å—Å–ª–µ–¥–æ–≤–∞–ª–∏ —Ö–∞—Ä–∞–∫—Ç–µ—Ä —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö. –ü–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è –∑–Ω–∞—á–∏–º—ã—Ö –≤—ã–±—Ä–æ—Å–æ–≤ –∏/–∏–ª–∏ –∞–Ω–æ–º–∞–ª–∏–π –Ω–µ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–æ.
- –ò—Å—Å–ª–µ–¥–æ–≤–∞–ª–∏ –ø—Ä–∏–∑–Ω–∞–∫–∏ –Ω–∞ –º—É–ª—å—Ç–∏–∫–æ–ª–ª–∏–Ω–µ–∞—Ä–Ω–æ—Å—Ç—å –∏ —É—Ç–µ—á–∫—É —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ —Å –ø–æ–º–æ—â—å—é –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ phik. –ò—Ç–æ–≥–æ –≤ —Å–≤—è–∑–∏ —Å –º—É–ª—å—Ç–∏–∫–æ–ª–ª–∏–Ω–µ–∞—Ä–Ω–æ—Å—Ç—å –∏–ª–∏ —Å —É—Ç–µ—á–∫–æ–π —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –±—ã–ª–∏ —É–±—Ä–∞–Ω—ã —Å–ª–µ–¥—É—é—â–∏–µ —Å—Ç–æ–ª–±—Ü—ã:

- –ü–æ–¥–≥–æ—Ç–æ–≤–∏–ª–∏ –¥–∞–Ω–Ω—ã–µ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –º–æ–¥–µ–ª–µ–π (–ø—Ä–æ–≤–µ–ª–∏ OHE –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ –∏ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ –¥–ª—è –º–æ–¥–µ–ª–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, –∞ —Ç–∞–∫–∂–µ Ordinal –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ –¥–ª—è –º–æ–¥–µ–ª–∏ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞).

–î–∞–Ω–Ω—ã–µ –ø–æ–¥–≥–æ—Ç–æ–≤–ª–µ–Ω—ã –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –∏ –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–µ–π.

## –û–±—É—á–µ–Ω–∏–µ –∏ –∞–Ω–∞–ª–∏–∑ —Ä–∞–±–æ—Ç—ã –º–æ–¥–µ–ª–µ–π

–í —Ç–µ—á–µ–Ω–∏–µ –¥–∞–Ω–Ω–æ–≥–æ —ç—Ç–∞–ø–∞ –Ω–∞–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ:
- –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –∏ –æ–±—É—á–∏—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ –º–æ–¥–µ–ª–µ–π (–º–∏–Ω–∏–º—É–º –¥–≤–µ), –∞ —Ç–∞–∫–∂–µ –ø–æ–¥–æ–±—Ä–∞—Ç—å –¥–ª—è –Ω–∏—Ö –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã. –î–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –±—ã–ª–∏ –≤—ã–±—Ä–∞–Ω—ã —Å–ª–µ–¥—É—é—â–∏–µ:
    - –°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å.
    - –õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è.
    - LGBM.
    - CatBoost.
- –û—Ü–µ–Ω–∫–∞ –º–µ—Ç—Ä–∏–∫ –º–æ–¥–µ–ª–µ–π. –í—ã–±–æ—Ä –ª—É—á—à–µ–π. –û—Ü–µ–Ω–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–µ–π –±—É–¥–µ—Ç –ø—Ä–æ–∏–∑–≤–¥–∏—Ç—å—Å—è —Å –ø–æ–º–æ—â—å—é –º–µ—Ç—Ä–∏–∫–∏ ROC AUC.
- –ü—Ä–æ–≤–µ—Ä–∫–∞ –≤—ã–±—Ä–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ. –ú–µ—Ç—Ä–∏–∫–∞ ROC AUC –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –Ω–µ –º–µ–Ω–µ–µ 0.85.

–ù–∞–ª–∏—á–∏–µ –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞ –∫–ª–∞—Å—Å–æ–≤ –≤ –Ω–∞—à–µ–º –¥–∞—Ç–∞—Å–µ—Ç–µ –±—É–¥–µ–º –Ω–µ–≤–µ–ª–∏—Ä–æ–≤–∞—Ç—å —Å –ø–æ–º–æ—â—å—é –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –º–æ–¥–µ–ª–µ–π.

```python
# –Ω–∞—á–Ω–µ–º —Å –º–æ–¥–µ–ª–∏ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞

random_forest = RandomForestClassifier(random_state=RANDOM_STATE)

pipline_forest = make_pipeline(col_transformer_oe, random_forest)

forest_parameters_grid = {
    
    # —É—á–∏—Ç—ã–≤–∞–µ–º –¥–∏—Å–±–∞–ª–∞–Ω—Å –∫–ª–∞—Å—Å–æ–≤ –≤ –Ω–∞—à–µ–º –¥–∞—Ç–∞—Å–µ—Ç–µ –∏ –ø—Ä–æ–≤–µ—Ä–∏–º –∫–∞—á–µ—Å—Ç–≤–æ —Ä–∞–±–æ—Ç—ã –º–æ–¥–µ–ª–∏ –ø—Ä–∏ –±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–µ –∫–ª–∞—Å—Å–æ–≤ –∏ –±–µ–∑ –Ω–µ—ë
    'randomforestclassifier__class_weight' : ['balanced', None],
    'randomforestclassifier__max_depth' : range(2, 7),
    'randomforestclassifier__n_estimators' : list(map(int, np.linspace(60, 160, 6)))
}

forest_grid_cv = GridSearchCV(
    pipline_forest,
    forest_parameters_grid,
    cv=5,
    scoring = 'roc_auc',
    verbose=3,
    n_jobs=-1
)
forest_grid_cv.fit(features_train_oe, target_train)

forest_best_score = round(forest_grid_cv.best_score_, 2)

print(f'–ù–∞–∏–ª—É—á—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ max_depth: {forest_grid_cv.best_params_}') 
print(f'–ù–∞–∏–ª—É—á—à–∏–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å ROC AUC: {forest_best_score}')
```

–ú—ã –ø–æ–ª—É—á–∏–ª–∏, —á—Ç–æ –Ω–∞–∏–±–æ–ª—å—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ ROC AUC –¥–æ—Å—Ç–∏–≥–∞–µ—Ç—Å—è –ø—Ä–∏ —Å–ª–µ–¥—É—é—â–∏—Ö –∑–Ω–∞—á–µ–Ω–∏—è—Ö –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤:
- max_depth = 6
- n_estimators = 100
- ROC AUC = 0.82

```python
# –Ω–∞—á–Ω–µ–º —Å –º–æ–¥–µ–ª–∏ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞

logistic_regression = LogisticRegression(random_state=RANDOM_STATE)

pipline_log_reg = make_pipeline(col_transformer_ohe_scaler, logistic_regression)


regression_parameters_grid = {
    'logisticregression__class_weight' : ['balanced', None],
    'logisticregression__penalty' : ['l1', 'l2'],
    'logisticregression__C' : [0.5, 1.0, 3.0, 5.0, 7.0, 10.0],
    'logisticregression__solver' : ['liblinear']
}

regression_grid_cv = GridSearchCV(
    pipline_log_reg,
    regression_parameters_grid,
    cv=5,
    scoring = 'roc_auc',
    verbose=3,
    n_jobs=-1
)
regression_grid_cv.fit(features_train_ohe_scaled, target_train)

log_regression_best_score = round(regression_grid_cv.best_score_, 2)

print(f'–ù–∞–∏–ª—É—á—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ max_depth: {regression_grid_cv.best_params_}') 
print(f'–ù–∞–∏–ª—É—á—à–∏–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å ROC AUC: {log_regression_best_score}')
```

–î–ª—è –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏:
- –° = 7
- class_weight = 'balanced'
- solver = 'liblinear'
- ROC AUC = 0.76

```python
# –¢–µ–ø–µ—Ä—å –ø–µ—Ä–µ—Ö–æ–¥–∏–º –∫ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–º—É –±—É—Å—Ç–∏–Ω–≥—É —Å –ø–æ–º–æ—â—å—é LGBM

lgbm_model = LGBMClassifier(random_state=RANDOM_STATE)

pipline_lgbm = make_pipeline(col_transformer_oe, lgbm_model)

lgbm_parameters_grid = {
    'lgbmclassifier__class_weight' : ['balanced', None],
    'lgbmclassifier__max_depth' : range(2, 6),
    'lgbmclassifier__n_estimators' : list(map(int, np.linspace(1300, 1700, 5))),
    'lgbmclassifier__learning_rate': [0.1]
}

lgbm_grid_cv = GridSearchCV(
    pipline_lgbm,
    lgbm_parameters_grid,
    cv=5,
    scoring = 'roc_auc',
    verbose=1,
    n_jobs=-1
)
lgbm_grid_cv.fit(features_train_oe, target_train)

lgbm_best_score = round(lgbm_grid_cv.best_score_, 2)

print(f'–ù–∞–∏–ª—É—á—à–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤: {lgbm_grid_cv.best_params_}') 
print(f'–ù–∞–∏–ª—É—á—à–∏–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å ROC AUC: {lgbm_best_score}')
```

```python
# –ø–µ—Ä–µ–π–¥–µ–º –∫ –ø–æ—Å–ª–µ–¥–Ω–µ–π –º–æ–¥–µ–ª–∏ - CatBoost
catboost_model = CatBoostClassifier(random_state=RANDOM_STATE, verbose=False)

pipline_catboost = make_pipeline(col_transformer_oe, catboost_model)

catboost_parameters_grid = {
    'catboostclassifier__max_depth' : range(2,6),
    'catboostclassifier__n_estimators' : list(map(int, np.linspace(1800, 2300, 6))),
    'catboostclassifier__learning_rate': [0.1]
}

catboost_grid_cv = GridSearchCV(
    pipline_catboost,
    catboost_parameters_grid,
    cv=5,
    scoring = 'roc_auc',
    verbose=1,
    n_jobs=-1
)
catboost_grid_cv.fit(features_train_oe, target_train)

catboost_best_score = round(catboost_grid_cv.best_score_, 2)

print(f'–ù–∞–∏–ª—É—á—à–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤: {catboost_grid_cv.best_params_}') 
print(f'–ù–∞–∏–ª—É—á—à–∏–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å ROC AUC: {catboost_best_score}')
```

–ò–º–µ–Ω–Ω–æ –µ—ë –º—ã –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ.

```python
catboost_model_test = CatBoostClassifier(max_depth=2, n_estimators=2300,
    learning_rate=0.1, random_state=RANDOM_STATE, verbose=False)

features_train_oe_final = pd.DataFrame(
    col_transformer_oe.fit_transform(features_train_oe),
    columns=col_transformer_oe.get_feature_names_out()
)

features_test_oe_final = pd.DataFrame(
    col_transformer_oe.transform(features_test_oe),
    columns=col_transformer_oe.get_feature_names_out()
)


catboost_model_test.fit(features_train_oe_final, target_train)

preds = catboost_model_test.predict(features_test_oe_final)

probabilities_test = catboost_model_test.predict_proba(features_test_oe_final)
probabilities_one_test = probabilities_test[:, 1]

best_roc_auc = roc_auc_score(target_test, probabilities_one_test)
print(f'–ó–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ ROC AUC –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ –¥–ª—è –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏: {best_roc_auc}')
```
